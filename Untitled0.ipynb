```python
#ques 1
import plotly.graph_objects as go
import seaborn as sns
tit=sns.load_dataset("titanic")
fig = go.Figure()
fig.add_trace(go.Scatter(x=tit.age , y =tit.fare))
fig.show()
```

```python
#Ques2
import plotly.express as px

# using the tips dataset
df = px.data.tips()

# plotting the box chart
fig = px.box(df,x="day" , y ="total_bill")

# showing the plot
fig.show()
```

```python
#Ques3
tips=sns.load_dataset("tips")
fig=go.Figure(data=[go.Histogram(x=tips.sex,y=tips.total_bill)])
fig.show()
```

```python
#Ques 4
import plotly.express as px
iris=sns.load_dataset("iris")
fig=px.scatter_matrix(iris,color="species")
fig.show()
```

```python
#Ques 5 what is distplot
#ans -The distplots in plotly python combines various statistical distribution such as histograms, violin plot, box plot etc in the same plot.

# plot a distplot using plotly express
import plotly.figure_factory as ff
import numpy as np

x1 = np.random.randn(200) - 1
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 1

hist_data = [x1, x2, x3]

group_labels = ['Group 1', 'Group 2', 'Group 3']
colors = ['#333F44', '#37AA9C', '#94F3E4']

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)
fig.show()
```
